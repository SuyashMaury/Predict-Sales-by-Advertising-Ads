```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
df['car_make']
```

```python
fig = px.bar(data_frame=df, x=df['car_make'].value_counts().index, y=df['car_make'].value_counts().values,text_auto=True,color=df['car_make'].value_counts().index)
fig.update_layout(title='Number of car brands',xaxis_title='cars_brands',yaxis_title='Count',width=800,height=500)
fig.show()
```

```python
fig = px.bar(data_frame=df, x=df['Fuel_Type'].value_counts().index, y=df['Fuel_Type'].value_counts().values,text_auto=True,color=df['Fuel_Type'].value_counts().index)
fig.update_layout(title='number of cars on the basis Fuel types',xaxis_title='fuels',yaxis_title='Count',width=800,height=500)
fig.show()
```

```python
fig=px.histogram(data_frame=df,x='Year',text_auto=True,nbins=31)
fig.update_layout(title='Number of cars sold per year',width=1000,height=500)
fig.show()
```
